## Task 2

```python
df_task2 = df[['category', 'repeat_purchase']]
df_task2
```

```python
df_task2 = df_task2.groupby(['category', 'repeat_purchase']).size()
df_task2
```

Create a visualization that shows how many products are repeate purchased has the most observations.

```python
# Reshape the data for plotting
df_plot = df_task2.unstack().fillna(0)

# Create a stacked bar chart
ax = df_plot.plot(kind='bar', stacked=True, figsize=(10, 6))

# Add labels and title
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Counts of Repeat Purchases by Category')

# Add a legend
plt.legend(title='Repeat Purchase', labels=['No', 'Yes'], loc='upper right')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
total_counts = df_task2.groupby('repeat_purchase').sum()
total_counts
```

In this case, "0" typically represents cases where there was no repeat purchase, and "1" represents cases where there was a repeat purchase. To assess balance or imbalance:

**Looking at the Total Counts:**

   - The total count of "0" values is 594.
   - The total count of "1" values is 906.

   There are more observations with "1" (repeat purchases) than "0" (non-repeat purchases) across all categories. This suggests an imbalance in favor of repeat purchases overall.

**Looking at Individual Categories:**

   When we examine the individual categories:
   - For "Accessory," there are 56 observations with "0" and 70 with "1," which is relatively balanced.
   - For "Equipment," there are 149 observations with "0" and 221 with "1," which is imbalanced in favor of repeat purchases.
   - For "Food," there are 109 observations with "0" and 151 with "1," which is also imbalanced in favor of repeat purchases.
   - For "Housing," there are 75 observations with "0" and 152 with "1," which is imbalanced in favor of repeat purchases.
   - For "Medicine," there are 84 observations with "0" and 153 with "1," which is imbalanced in favor of repeat purchases.
   - For "Toys," there are 110 observations with "0" and 145 with "1," which is relatively balanced.
   - For "Unknown," there are 11 observations with "0" and 14 with "1," which is also relatively balanced.

**Overall Assessment:**

   The imbalance or balance in the "repeat_purchase" variable across different categories depends on the specific category. Some categories are more balanced, while others are imbalanced in favor of repeat purchases. The total dataset is also imbalanced, as there are more "1" values (repeat purchases) than "0" values (non-repeat purchases) overall.

   Whether this imbalance is a concern depends on the context of your analysis and the specific goals of your project. ## Task 4

```python
df_task4 = df[['sales', 'repeat_purchase']]
df_task4
```

```python
sale_counts = df_task4.groupby('repeat_purchase').sum()
sale_counts
```

Bar chart of the `category` and `repeat_purchase` columns

```python
repeat_purchase_counts = sale_counts.index
sales_0 = sale_counts.loc[0]['sales']
sales_1 = sale_counts.loc[1]['sales']

# Create a stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(repeat_purchase_counts, [
        sales_0, sales_1], color=['skyblue', 'lightcoral'])
plt.xlabel('Repeat Purchase')
plt.ylabel('Total Sales')
plt.title('Total Sales by Repeat Purchase')
plt.xticks(repeat_purchase_counts, ['No', 'Yes'])
plt.show()
```

Based on the provided data and calculations:

- For products with a "repeat_purchase" value of 0, the total sales amount is $610,850.60.
- For products with a "repeat_purchase" value of 1, the total sales amount is $884,046.17.

We can observe the following relationship between repeat purchase and sales:

Products with a "repeat_purchase" value of 1 (meaning they were repeat-purchased) have a higher total sales amount ($884,046.17) compared to products with a "repeat_purchase" value of 0 ($610,850.60). This suggests that products that are more likely to be repeat-purchased tend to generate higher sales revenue.

In other words, there appears to be a positive relationship between repeat purchases and sales, indicating that products that customers are more likely to repurchase contribute more to the total sales revenue. This information could be valuable for understanding customer behavior and making decisions related to product offerings and marketing strategies.
